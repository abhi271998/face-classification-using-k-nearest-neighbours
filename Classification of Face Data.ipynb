```python
import cv2
import numpy as np
import os
```

```python
def dist(x, y):
    return np.sqrt(((x - y)**2).sum())

def KNN(X_data, Y_data, query, k):
    
    points = []
    for ix in range(0, X_data.shape[0]):
        d = dist(query, X_data[ix])
        points.append((d, Y_data[ix]))
    
    points = sorted(points)
    points = np.array(points)
    arr = points[:k, 1]
    t = np.unique(arr, return_counts = True)
    index = t[1].argmax()
    pred = t[0][index]
    return pred
```

```python
X_data = []
Y_data = []
class_id = 0
for file in os.listdir('./data/'):
    print(file)
    data_item = np.load('./data/' + file)
    print(data_item.shape)
    X_data.append(data_item) # 44 images
    Y_data.append(class_id * np.ones((data_item.shape[0], )))
    class_id += 1
    
X_data = np.array(X_data)
Y_data = np.array(Y_data)

face_data_x = np.concatenate(X_data, axis = 0)
face_data_y = np.concatenate(Y_data, axis=0).reshape((-1,1))
print(X_data.shape, Y_data.shape, face_data_x.shape, face_data_y.shape)
```

```python
cap = cv2.VideoCapture(0)
dataset_path = './data/'
k = 0
dic={0:"abhishek",1:"Roli di",2:"mummy",3:"papa"}
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_alt.xml')
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    if ret == False:
        continue
    
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for face in faces:
        x,y,w,h = face
        offset = 5
        face_section = frame[y-offset : y + h + offset, x-offset:x + w + offset]
        face_section = cv2.resize(face_section, (100,100))
        
        predicition = KNN(face_data_x, face_data_y, face_section.flatten(), 5)
        
        
#         if predicition==in range(0-4)
        cv2.putText(frame, str(dic[predicition[0]]) , (x, y - 10), cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0), 2, cv2.LINE_AA)
        
        cv2.rectangle(frame, (x-offset,y-offset), (x+w+offset, y+h+offset), (255,0,0), 2)
        
    
    k = k+1
    print(k)
    cv2.imshow("Face", frame)
   
    if cv2.waitKey(1) & 0xFF == ord('r'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
